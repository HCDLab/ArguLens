```python
import pandas as pd
import requests 
import sys
import json

# https://stanfordnlp.github.io/CoreNLP/corenlp-server.html
```

```python
# run the following comment to host the CoreNLP server first
# java -mx4g -cp "*" edu.stanford.nlp.pipeline.StanfordCoreNLPServer -port 9000 -timeout 15000

# api-endpoint 
URL = "http://localhost:9000/"
```

```python
df = pd.read_csv('Conversational_data.csv')
```

```python
# defining a params dict for the parameters to be sent to the API 
PARAMS = {'properties': "{'annotators': 'tokenize,ssplit,pos,parse,depparse'}"}
  
def GetPos(row):
    text = row['Text Content']
    pos_list = []
    try:
        response = requests.post(
            URL, params=PARAMS,
            data=text.encode('UTF-8')
        )
        response.raise_for_status()
        for sentence in response.json()['sentences']:
            for token in sentence['tokens']:
                pos = token['pos']
                pos_list.append(pos)
    except Exception as e:
        print(e)
    pos_str = ' '.join(pos_list)
    print(pos_str)
    return pos_str
```

```python
df['pos_str'] = df.apply(GetPos, axis=1)
```
