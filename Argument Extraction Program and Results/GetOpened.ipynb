```python
import re
import requests
import pandas as pd
from datetime import datetime, timedelta
```

```python
datetime_format_github = '%Y-%m-%dT%H:%M:%SZ'
```

```python
# get author_association
access_token = 'xxx'  # removed access token for anonymity

def GetOpened():
    created_time_list = []
    for page_num in range(1,3000):
        if page_num%10 == 0:
            print(page_num)
        request_str = "https://api.github.com/repos/microsoft/vscode/issues?direction=asc&state=all&page=" + str(page_num) + "&access_token=" + access_token
#         print(request_str)
        r = requests.get(request_str)
        try:
            createdList = r.json()
#             print(commitList)
            if len(createdList) == 0:
                break
            for issue in createdList:
                created_time = issue['created_at']
                created_time_list.append(created_time)
            
        except:
            print('EXCEED ACCESS LIMIT')
    return created_time_list
```

```python
df = pd.DataFrame(GetOpened(), columns=['datetime_str']) 
```

```python
df['datetime'] = df.apply(lambda row: datetime.strptime(row['datetime_str'], datetime_format_github), axis=1)
```

```python
start_date_time = datetime.strptime('2015-09-03T20:23:38Z', datetime_format_github)
```

```python
def GetDiffWeek(d1, d2):
    monday1 = (d1 - timedelta(days=d1.weekday()))
    monday2 = (d2 - timedelta(days=d2.weekday()))
    return int((monday2 - monday1).days / 7)
```

```python
df['week_diff'] = df.apply(lambda row: GetDiffWeek(start_date_time, row['datetime']), axis=1)
```
